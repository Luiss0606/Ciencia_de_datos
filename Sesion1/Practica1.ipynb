**Clasificacion Binaria:**

```python
from numpy import where
from collections import Counter
from sklearn.datasets import make_blobs
from matplotlib import pyplot

#Utilizamos la función make_blobs() para generar un conjunto de datos de clasificación binaria sintética.
X, y = make_blobs(n_samples=8000, centers=4, random_state=1)

print(X.shape, y.shape)

#Cuenta los elementos de y
counter = Counter(y)
print(counter)
print(counter.items())

#Pintamos los 10 primeros ejemplos -loop
for i in range(10):
  {
		print(X[i], y[i])
  }
# Graficamos el conjunto de datos y coloreamos segun la etiqueta de la clase
for label, _ in counter.items():
	print(label)

	row_ix = where(y == label)[0]
	#print(X[row_ix, 0])
	#print(X[row_ix, 1])
	if label == 0:
		pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label), c="yellow")
	elif label == 1:
		pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label), c="red")
	elif label == 2:
		pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label), c="blue")
	else:
		pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label), c="green")
pyplot.legend()
pyplot.show()
```

**Clasificacion Multi Clase**

```python
from numpy import where
from collections import Counter
from sklearn.datasets import make_blobs
from matplotlib import pyplot

X, y = make_blobs(n_samples=10000, centers=3, random_state=1)

print(X.shape, y.shape)

counter = Counter(y)
print(counter)

for i in range(10):
  print(X[i], y[i])

for label, _ in counter.items():
  row_ix = where(y == label)[0]
  pyplot.scatter(X[row_ix, 0], X[row_ix, 1], label=str(label))
pyplot.legend()
pyplot.show()
```
